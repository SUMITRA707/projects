```python
import fitz  # PyMuPDF
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files
import io

```

```python
def extract_text_from_pdf(pdf_file):
    doc = fitz.open(stream=pdf_file.read(), filetype="pdf")
    text = ""
    for page in doc:
        text += page.get_text()
    return text
```

```python
def calculate_similarity(resume_texts, job_description):
    documents = [job_description] + resume_texts
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(documents)
    cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()
    return cosine_sim
```

```python
job_description = input("Paste the job description here:\n")

```

```python
print("Upload resume PDFs:")
uploaded_files = files.upload()

resume_texts = []
filenames = []

for filename, filedata in uploaded_files.items():
    text = extract_text_from_pdf(io.BytesIO(filedata))
    resume_texts.append(text)
    filenames.append(filename)
```

```python
scores = calculate_similarity(resume_texts, job_description)

results_df = pd.DataFrame({
    'Resume File': filenames,
    'Matching Score (%)': [round(s * 100, 2) for s in scores]
}).sort_values(by='Matching Score (%)', ascending=False)

results_df.reset_index(drop=True, inplace=True)
results_df
```

```python
results_df.to_csv("ranked_resumes.csv", index=False)
files.download("ranked_resumes.csv")
```
